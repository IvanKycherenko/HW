```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x, y):
    return np.cos(x + 0.5) - y - 2

def f2(x, y):
    return np.sin(y) - 2 * x - 1

x_values = np.linspace(-2, 2, 400)
y_values = np.linspace(-2, 2, 400)

X, Y = np.meshgrid(x_values, y_values)
Z1 = f1(X, Y)
Z2 = f2(X, Y)

plt.contour(X, Y, Z1, levels=[0], colors='r', label='f1(x, y) = 0')
plt.contour(X, Y, Z2, levels=[0], colors='b', label='f2(x, y) = 0')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Графічне відокремлення коренів системи рівнянь')
plt.grid(True)
plt.show()
```

```python
# Функції для методу простої ітерації
def g1(x, y):
    return np.cos(x + 0.5) - 2

def g2(x, y):
    return np.arcsin(2 * x + 1)

# Початкові наближені значення
x0, y0 = 0, 0

# Ітерації методу простої ітерації
for i in range(1, 6):
    x0 = g1(x0, y0)
    y0 = g2(x0, y0)
    print(f"Iтерація {i}: x = {x0:.5f}, y = {y0:.5f}")
```
