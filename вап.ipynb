```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, x0, y0, a, b, h):
    n = int((b - a) / h) + 1
    x_values = np.linspace(a, b, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_values[i] = y_values[i-1] + h * f(x_values[i-1], y_values[i-1])

    return x_values, y_values

def f(x, y):
    return x + np.cos(y / np.sqrt(3))

x0, y0 = 1.2, 2.1

x_values, y_values = euler_method(f, x0, y0, 1.2, 2.2, 0.1)

plt.plot(x_values, y_values, label='Метод Ейлера')

plt.title('Ламана Ейлера для розв\'язку')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_cauchy_method(f, x0, y0, a, b, h):
    n = int((b - a) / h) + 1
    x_values = np.linspace(a, b, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_temp = y_values[i-1] + h * f(x_values[i-1], y_values[i-1])
        y_values[i] = y_values[i-1] + 0.5 * (f(x_values[i-1], y_values[i-1]) + f(x_values[i], y_temp)) * h

    return x_values, y_values

def f(x, y):
    return x + np.sin(y / np.sqrt(2))

x0, y0 = 0.8, 1.3

x_values, y_values = euler_cauchy_method(f, x0, y0, 0.8, 1.8, 0.1)

plt.plot(x_values, y_values, label='Метод Ейлера-Коші')

plt.title('Ламана Ейлера для розв\'язку')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
