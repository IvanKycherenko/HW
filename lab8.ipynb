```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x_data = np.array([0.7, 0.8, 1.3, 1.5, 2.1])
y_data = np.array([2.63, 2.87, 2.19, 1.76, 3.43])

cs = CubicSpline(x_data, y_data)

x_interp = np.linspace(min(x_data), max(x_data), 1000)
y_interp = cs(x_interp)

plt.scatter(x_data, y_data, label='Таблична функція')
plt.plot(x_interp, y_interp, label='Кубічний сплайн', color='red')
plt.title('Кубічний сплайн для таблично заданої функції')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
